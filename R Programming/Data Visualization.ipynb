```r
# Example
ggplot(diamonds, aes(price)) +
  geom_histogram(fill="salmon",
                 bins=100) +
  theme_minimal()
```

```r
# Check data
glimpse(diamonds)
summary(diamonds)
```

## Chart 1 
Relationship between carat and price group by cut

```r
ggplot(sample_n(diamonds, 1000), aes(price, carat, color=color)) +
  geom_point(alpha = 0.3) +
  geom_smooth(method = lm, formula = y ~ x) +
  theme_minimal() +
  facet_wrap(~ color, ncol = 4)
  labs(title = "Relationship between carat and price group by color",
        x= "Price",
        y= "Carat")
```

## Chart 2
Number of diamonds by color and cut

```r
ggplot(diamonds, aes(y = color, fill = cut)) +
  geom_bar() +
  theme_minimal() +
  labs(title = "Number of diamonds by color and cut",
        x= "Amount",
        y= "Color")
```

## Chart 3
Relationship between carat and cut

```r
ggplot(diamonds, aes(carat,cut,fill=cut)) +
  geom_boxplot() +
  theme_minimal() +
  labs(title = "Relationship between carat and cut",
        x= "Carat",
        y= "Cut")
```

## Chart 4
Relationship between depth and cut

```r
ggplot(diamonds, aes(depth,cut, fill=cut)) +
  geom_violin() +
  theme_minimal() +
  labs(title = "Relationship between depth and cut",
        x= "Depth",
        y= "Cut")
```

## Chart 5
Relationship between depth and cut

```r
ggplot(diamonds, aes(price,depth,color=color)) +
  geom_point(size=2, alpha=0.3) +
  theme_minimal() +
  facet_wrap(~ clarity)
  labs(title = "Relationship between carat and cut",
        x= "Depth",
        y= "Price")
```
